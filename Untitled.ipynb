```python
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
from PIL import Image
%matplotlib inline

# fid-score 
fid_scores = np.loadtxt('fid_scores.csv')
fid_scoresW = np.loadtxt('fid_scoresW.csv')

fid_scores = fid_scores.tolist()
fid_scoresW = fid_scoresW.tolist()
```

```python
##Visualize FID_score vs epochs
fig, ax = plt.subplots()
plt.plot(fid_scores, label='DCGAN', alpha=1)
plt.plot(fid_scoresW, label='WGAN-GP', alpha=1)
plt.title("FID scores vs epoch", fontsize=16, pad=16)
plt.xlabel("epochs")
plt.ylabel("fid-score")
plt.legend()
```

```python
# Importing Image class from PIL module
from PIL import Image

# Opens a image in RGB mode
img1 = Image.open("./report/dcgan1.png")
img50 = Image.open("./report/dcgan50.png")

img1w = Image.open("./report/wgan1.png")
img50w = Image.open("./report/wgan50.png")

# Size of the image in pixels (size of original image)
# (This is not mandatory)
width, height = img1.size

# Setting the points for cropped image
left = 1
top = 1
right = width / 2
bottom = height / 4 + 1

# Cropped image of above dimension
# (It will not change original image)
dc1 = img1.crop((left, top, right, bottom))
dc50 = img50.crop((left, top, right, bottom))
w1 = img1w.crop((left, top, right, bottom))
w50 = img50w.crop((left, top, right, bottom))

# Shows the image in image viewer
dc1.save("dc1.png")
dc50.save("dc50.png")
w1.save("w1.png")
w50.save("w50.png")
```
